## how to tain a agent to play CartPole-v1 by using Tao

```python
# train "CartPole-v1" model
agent = PPO(env_id="CartPole-v1", total_timesteps=500000, track=False, capture_video=True)
agent.learn()
```

```python
# save model
import torch
is_save = True
if is_save:
    torch.save(agent.state_dict(), "./ppo.pt")
    
# load model
model = PPO(env_id="CartPole-v1")
model.load_state_dict(torch.load("./ppo.pt", map_location="cpu"))
model.eval()
```

```python
# log huggingface account
from huggingface_hub import notebook_login
notebook_login()
```

```python
# push "CartPole-v1" model to hugging face
from tao.utils.metrics import evaluate
from tao.utils.huggingface import push_to_hub

episodic_returns = evaluate(model, eval_episodes=10, epsilon=0.05)
repo_id = "amu/tao-ppo-CartPole-v1" # change it to your repo name
local_repo_path="hub"
push_to_hub(model, repo_id, local_repo_path, episodic_returns)
```
